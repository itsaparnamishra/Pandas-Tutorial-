```python
import numpy as np
import pandas as pd
```

```python
#simple graph with x(0,0) and y(0,625)
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = np.array([0,6])
y = np.array([0,625])
plt.plot(x,y)
plt.show()
```

```python
#x(1,8) and y(3,10) 
x = np.array([1,3])
y = np.array([8,10])
plt.plot(x,y,"o") # "o" is used to mark the points
plt.show()
```

```python
#Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
#plotting just the points 

x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o")
plt.show()
```

```python
# if the x points are not specified then the default values assigned will be 0,1,2,3,4 etc
# depending on the length of y points 

#marker argument is used to specify the points 
y = np.array([7,8,9,0,2,3,5])
plt.plot(y,marker = "o")
plt.show()
```

```python
# marker can also be "*"
y = np.array([7,8,9,0,2,3,5])
plt.plot(y,marker = "*")
plt.show()
```

```python
# marker ="-"
y = np.array([7,8,9,0,2,3,5])
plt.plot(y,marker = ".")
plt.show()
```

```python
# format strings ----- fmt

y = np.array([7,8,9,0,2,3,5])
plt.plot(y,"o:g")   # "o:g" o is the marker and g is green color 
plt.show()
```

```python
# markersize ----- ms 

y = np.array([7,8,9,0,2,3,5])
plt.plot(y,marker = "o",ms = 9)
plt.show()
```

```python
# linestyle ----- for "dotted" lines

y = np.array([7,8,9,0,2,3,5])
plt.plot(y ,marker = "o" ,linestyle = "dotted")
plt.show()
```

```python
# linestyle ------ for "dashed" lines 
y = np.array([7,8,9,0,2,3,5])
plt.plot(y,marker = "o",linestyle="dashed",color = "r")
plt.show()
```

```python
# multiple lines 


y1 = np.array([7,8,9,0,2,3,5])
y2 = np.array([1,2,3,4,5,6,7])

plt.plot(y1,marker = "o")
plt.plot(y2,marker = "*")

plt.show()
```

```python
# labels ------ xlabel : for x axis
#               ylabel : for y axis

x = np.array([50,55,60,65,70,75,80])
y = np.array([20,24,36,39,45,40,59])
plt.plot(x,y,"o",linestyle = "dashed")


plt.title("Tournament score data",loc = "right")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
# grid() ------ this function is used to add grid lines to the plot 


x = np.array([33,44,55,66,77,88,99,100])
y = np.array([111,222,333,444,555,666,777,888])
plt.plot(x,y,"o",linestyle="dashed")

plt.title("Graph",loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid()

plt.show()
```

```python
#  we can also specify which grid lines to display 

x = np.array([33,44,55,66,77,88,99,100])
y = np.array([111,222,333,444,555,666,777,888])
plt.plot(x,y,"o",linestyle="dashed")

plt.title("Graph",loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(axis= "y")

plt.show()
```

```python
x = np.array([33,44,55,66,77,88,99,100])
y = np.array([111,222,333,444,555,666,777,888])
plt.plot(x,y,"o",linestyle="dashed")

plt.title("Graph",loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(axis= "x")

plt.show()
```

```python
#we can also add properties like color,linestyle, linewidth inside the grid function 

x = np.array([33,44,55,66,77,88,99,100])
y = np.array([111,222,333,444,555,666,777,888])
plt.plot(x,y,"o",linestyle="dashed")

plt.title("Graph",loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(color ="red",linestyle="--",linewidth=0.8,axis= "y")

plt.show()
```

```python
# subplots ------subplot() function can help us drwa multiple plots in one figure 
#takes three arguments ------plt.subplot(rows,cols,index)

#plot 1
x1 = np.array([0,1,2,3,4])
y1 = np.array([3,6,8,2,9])

plt.subplot(1,2,1)   ##(1,2,1) where 1=no. of rows,2= no. of cols,index=1
plt.plot(x1,y1)
plt.title("My income")


#plot 2

x2 = np.array([15,16,17,18,19])
y2 = np.array([19,18,17,16,15])

plt.subplot(1,2,2)   #(1,2,2) where 1=no. of rows,2= no. of cols,index=2
plt.plot(x2,y2)
plt.title("My Sales")

plt.suptitle("My Shop")
plt.show()
```

```python
#colormap-----cmap




x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y,c=colors,cmap="viridis")


plt.colorbar()   #we can aslo include the colormap by using plt.colorbar() statement
plt.show()
```

```python
# bar graphs ----------------- bar()

x =np.array(["A","B","C","D","E"])
y = np.array([3,4,5,6,7])

plt.bar(x,y,color="r")

plt.show()
```

```python
# horizontal bar graphs -------------- barh()

x =np.array(["A","B","C","D","E","F"])
y = np.array([3,4,5,6,7,1])

plt.barh(x,y,color="g")

plt.show()
```

```python
#barwidth can aslo be added to get a specific width of the bar graph 

x = np.array(["Audi","Benz","Jaguar","Ford","Ferrari"])
y = np.array([123,45,678,98,333])
plt.xlabel("Cars")
plt.ylabel("Income(total) from Jan to June")

plt.bar(x,y,color="cyan",width=0.9)
plt.show()
```

```python
#Histograms ----------- hist()

# are used for frequency distributions 

x = np.random.normal(180,10,360)

plt.hist(x)
plt.show()
```

```python
# Pie charts

y = np.array([10,67,45,9])
labels = ["Benz","Audi","Ford","Ferrari"] 



plt.pie(y,labels=labels)
plt.show()
```

```python
# startangle and explode 


y = np.array([10,67,45,9])
labels = ["Benz","Audi","Ford","Ferrari"] 
explode = [0.3,0.3,0,0]


plt.pie(y,labels=labels,startangle=90,explode = explode)
plt.show()
```

```python
# shadow ----- we can also add a shadow parameter
#colors can also be added by specifying each colors for every section of the pie chart

y = np.array([10,67,45,9])
labels = ["Benz","Audi","Ford","Ferrari"] 
explode = [0.3,0.3,0,0]
color = ["r","g","b","hotpink"]


plt.pie(y,labels=labels,startangle=90,explode = explode,shadow=True,colors=color)
plt.show()
```

```python
# to add a list of explanation for each wedge , using legend() function 

y = np.array([10,67,45,9])
labels = ["Benz","Audi","Ford","Ferrari"] 
explode = [0.3,0.3,0,0]
color = ["r","g","b","hotpink"]


plt.pie(y,labels=labels,startangle=90,explode = explode,shadow=True,colors=color)
plt.legend(title ="Sales of Cars")
plt.show()
```
